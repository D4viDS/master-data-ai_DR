# Support Vector Machine

Al igual que vimos antes, el algoritmo SVM nos ayuda a clasificar observaciones en distintas categorías encontrando la frontera más grande entre ambas categorías (support vector). Ya vimos en la presentación qué es lo que hace, así que implementémoslo sobre la misma base de spotify que teníamos, y luego probemos con alguna otra base de datos.

```python
import numpy as np
import pandas as pd
```

```python
spotify = pd.read_csv('data_spotify.csv')
spotify = spotify.iloc[:, 1:]
spotify.head()
```

```python
spotify.describe()
```

```python
spotify.shape
```

```python
X = spotify.iloc[:, :13]
X.head()
```

```python
y = spotify.target
y.head()
```

Recuerden que es importante que reescalemos nuestras variables! Pero primero hay que dividir nuestra base

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
print(X_train)
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
print(X_train)
```

```python
from sklearn.svm import SVC

svclassifier = SVC(kernel='linear')

svclassifier.fit(X_train, y_train) 
```

```python
y_pred = svclassifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred)) 
```

```python
accuracy = accuracy_score(y_test, y_pred)
print('DecisionTreeClassifier accuracy score: {}'.format(accuracy))
```

## Probemos si no es porque la solución no es lineal

```python
svclassifier = SVC(kernel='poly', degree = 3)

svclassifier.fit(X_train, y_train) 
```

```python
y_pred = svclassifier.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))
```

**¿Mejoró**, pero esto qué me dice? ¿Qué hay de la C?

```python
svclassifier = SVC(kernel='poly', degree = 3, C=3)

svclassifier.fit(X_train, y_train) 
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

Recuerden que lo que hicimos al aumentar la C de 1 a 3 lo que sucedió fue que la "calle" se hizo más angosta y con ello permitió menos violaciones a la frontera. ¿Qué problemas tiene esto? ¿Qué ventajas?

Probemos también otros kernels, no sólo el que "agrega" polinomios

```python
svclassifier = SVC(kernel='rbf') # kernel gaussiano

svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
svclassifier = SVC(kernel='sigmoid') # kernel sigmoid

svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_test)

accuracy_score(y_test, y_pred)
```

## Parece que el gaussiano funciono. Por?

```python
svclassifier = SVC(kernel='rbf') # kernel gaussiano

svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_test)

accuracy_score(y_test, y_pred)
```

### Estos claramente no son nuestros datos, porque nosotros netemos N dimensiones, pero es para entender qué pasa. El kernel gaussiano lo que hace es poner los datos en una distribución Gaussiana (como de campana, normal) para separar las categorías.
