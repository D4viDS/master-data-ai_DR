## Booleanos y condicionales

## Booleanos

Los booleanos es un tipo de datos que represantan uno de dos valores, `True` (verdadero) o `False` (falso)

```python
# De la siguiente forma los podemos asignar a una variable
x = True
y = False
print(x, y)
```

```python
# bool() convierte una variable en booleano
print(bool('Text random'))
```

```python
# En python la mayoria de los valores son verdaderos
print(bool("abc"))
print(bool(123))
print(bool(["aguacate", "banana", "cereza"]))
```

```python
# Pero strigns vacios, el numero 0 y listas vacias son falsos
print(bool(""))
print(bool(0))
print(bool([]))
```

### Operadores de comparacion

|Operator|Name|Example|
|--- |--- |--- |
|==|igual a|x == y|
|!=|no igual a|x != y|
|>|mayor que|x > y|
|<|menor que|x < y|
|>=|igual o mayor que|x >= y|
|<=|igual o menor que|x <= y|

```python
a = 200
b = 30
c = 30

print("a == b", a == b) # False
print("b == c", b == c) # True
print("b != c", b != c) # F
print("a > b", a > b) # True
print("a < b", a < b) # False
print("b >= c", b >= c) # True
print("b <= c", b <= c) # True
```

### Operadores logicos

|Operator|Description|Example|
|--- |--- |--- |
|and|Verdadero si ambos valores son verdaderos|x < 5 and  x < 10| 
|or|Verdadero si uno de los valores es verdaderos|x < 5 or x < 4| 
|not|Verdadero si es falso, falso si es verdadero|not(x < 5 and x < 10)|
|&|AND|x < 5 & x < 10|
|\||OR|x < 5 \| x < 4|
|~|NOT|~(x < 5 & x < 10)|




```python
imc = 7
print("x < 5 and x < 10", imc >= 18.5 and imc < 24.9) # False y True. -> False
print("x < 5 or x < 10", x < 5 or x < 10) # False o True - > True
print("not(x < 5)", not(x > 5))
```

### Operadores de pertenencia

|Operator|Description|Example|
|--- |--- |--- |
|in|Verdadero si una secuencia con el valor especificado esta persente en el objeto|x in y|
|not in|Verdadero si una secuencia con el valor especificado no esta persente en el objeto|x not in y|


```python
print("[0] in [0, 1, 2]" , [0] in [0, 1, 2])
print("[3] in [0, 1, 2]" , [3] in [0, 1, 2])
print("'b' in 'banana'", "b" in "banana")
print("'c' in 'banana'", "b" in "banana")
```

## Condicionales


Los booleanos y operadores booleanos son principalmente usados en condicionales `if` `elif` y `else` y en ciclos `while`

### Condicionales if

Un condicional `if` permite escribir un segmento de codigo que solo se ejecute cundo se cumpla cierta condición

```python
if False:
  print('Esto se imprime por ser verdadero')
else:
  print('Esto no se imprime por ser falso')
```

```python
# Lo util de esto es que podemos poner expresiones booleanas
x = 5
y = 10
if x > y:
  print("x es mayor a y")
else: 
  print('Falso')


if ...:
  print("x es menor a y")
```

Para poder usar esto, es importante aprender la sintaxis (la forma correcta de escribir) de Python

```python
# En python, se hace uso de la indentacion para indicar que cierto segmento de
#   codigo es parte de alguna condicional

if True:
  print("esto es parte del condicional if")

print("esto no lo es")
```
