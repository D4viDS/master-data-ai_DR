#Probabilidad


* l, m y n representan tres probabilidades individuales.
* p representa el número de ensayos (hasta 100)
* q, r y s representan las probabilidades binomiales acumuladas, es decir, el aumento de la probabilidad por cada unidad de aumento del número de ensayos


En la teoría de la probabilidad y la estadística, una distribución de probabilidad es la función matemática que da las probabilidades de ocurrencia de los diferentes resultados posibles para un experimento


```python
l = 0.02
m = 0.04
n = 0.06
p=np.arange(0, 100, 1)
h = 1 - l
j = 1 - m
k = 1 - n
q = 1-(h**p)
r = 1-(j**p)
s = 1-(k**p)
```

Vemos que para las probabilidades q, r y s - las probabilidades acumuladas aumentan a ritmos diferentes para un número determinado de ensayos.

#Bayes

```python
# calculate P(A|B) given P(A), P(B|A), P(B|not A)
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
	# calculate P(not A)
	not_a = 1 - p_a
	# calculate P(B)
	p_b = p_b_given_a * p_a + p_b_given_not_a * not_a
	# calculate P(A|B)
	p_a_given_b = (p_b_given_a * p_a) / p_b
	return p_a_given_b
 
# P(A)
p_a = 0.0002
# P(B|A)
p_b_given_a = 0.85
# P(B|not A)
p_b_given_not_a = 0.05
# calculate P(A|B)
result = bayes_theorem(p_a, p_b_given_a, p_b_given_not_a)
# summarize
print('P(A|B) = %.3f%%' % (result * 100))
```

#Otras distribuciones

```python
X = stats.binom(10, 0.2) # Declare X to be a binomial random variable
print(X.pmf(3))           # P(X = 3)
print(X.cdf(4))           # P(X <= 4)
print(X.mean())           # E[X]
print(X.var())            # Var(X)
print(X.std())            # Std(X)
print(X.rvs())            # Get a random sample from X
print(X.rvs(10)) 
```

```python
B = stats.expon(4)    # Declare B to be an exponential random variable
print(B.pdf(1))       # f(1), the probability density at 1
print(B.cdf(2))       # F(2) which is also P(B < 2)
print(B.rvs()) 
```

```python
Y = stats.poisson(2)  # Declare Y to be a poisson random variable
print(Y.pmf(3))       # P(Y = 3)
print(Y.rvs())
```

```python
X = stats.geom(0.75)  # Declare X to be a geometric random variable
print(X.pmf(3))       # P(X = 3)
print(X.rvs()) 
```

```python
X = stats.beta(1, 3)  # Declare X to be a beta random variable
print(X.pdf(0.5))     # f(0.5), the probability density at 1
print(X.cdf(0.7))     # F(0.7) which is also P(X < 0.7)
print(X.rvs()) 
```
