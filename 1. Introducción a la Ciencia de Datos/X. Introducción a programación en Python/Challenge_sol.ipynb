## Formato de cadenas de texto

**f-Strings**: una forma nueva y mejorada de formatear cadenas en Python

Se unieron a la fiesta en Python 3.6. Las f-strings son **cadenas** literales que tienen una **f al principio** y **llaves** que contienen expresiones que serán reemplazadas por sus **valores**

**%-formatting:** Aquí hay algunos especificadores de argumentos básicos a conocer:
- `%s`: cadena (o cualquier objeto con una representación de cadena, como números) 
- %d: enteros
- `%f`: números de coma flotante
- `%.<número de dígitos>f`: números de coma flotante con una cantidad fija de dígitos a la derecha del punto.
- `%x/%X`: números enteros en representación hexadecimal (minúsculas / mayúsculas)

*una vez que comience a usar varios parámetros y cadenas más largas, su código rápidamente se volverá mucho menos legible.*


En este ejercicio deberá **escribir una cadena** de formato que **imprima** los datos utilizando la siguiente oracion: `Hola John Doe. Tu saldo actual es 53.44`

```python
data = ["John", "Doe", 53.44]
format_string = f"Hola {data[0]} {data[1]}. Tu saldo actual es de {data[2]}"

print(format_string)
```

```python
data = ("John", "Doe", 53.44)
format_string = "Hola %s %s. Tu saldo actual es de %.2f"

#print(format_string % ("John", "Doe", 53.44))
print(format_string % data)
```

## Funciones

En este ejercicio, utilizará una función existente y, al mismo tiempo, agregará la suya para crear un programa completamente funcional.
1. Agregue una función llamada list_benefits() que devuelva la siguiente lista de cadenas: "Código más organizado", "Código más legible", "Reutilización de código más fácil", "Permitir a los programadores compartir y conectar código juntos".
2. Agregue una función llamada build_sentence(info) que recibe un solo argumento que contiene una cadena y devuelve una oración que comienza con la cadena dada y termina con la cadena "¡es un beneficio de las funciones!"
3. ¡Ejecute y vea cómo todas las funciones funcionan juntas!

```python
# Modifica esta función para devolver una lista de cadenas de texto (las del punto 1)
def list_benefits():
    return ["Código más organizado", "Código más legible", "Reutilización de código más fácil", "Permitir a los programadores compartir y conectar código juntos"]

# Modifica esta función para concatenar a cada benefit (la cadena de texto) " es un beneficio de las funciones!"
def build_sentence(benefit):
    s = benefit + " ¡es un beneficio de las funciones!"
    return s


#probemos nuestras funciones definidas
def name_the_benefits_of_functions():
    list_of_benefits = list_benefits()
    for benefit in list_of_benefits:
        print(build_sentence(benefit))

#mandamos llamar a la función 
name_the_benefits_of_functions()
```
