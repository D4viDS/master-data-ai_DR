## Formato de cadenas de texto

**f-Strings**: una forma nueva y mejorada de formatear cadenas en Python

Se unieron a la fiesta en Python 3.6. Las f-strings son **cadenas** literales que tienen una **f al principio** y **llaves** que contienen expresiones que serán reemplazadas por sus **valores**

**%-formatting:** Aquí hay algunos especificadores de argumentos básicos a conocer:
- `%s`: cadena (o cualquier objeto con una representación de cadena, como números) 
- %d: enteros
- `%f`: números de coma flotante
- `%.<número de dígitos>f`: números de coma flotante con una cantidad fija de dígitos a la derecha del punto.
- `%x/%X`: números enteros en representación hexadecimal (minúsculas / mayúsculas)

*una vez que comience a usar varios parámetros y cadenas más largas, su código rápidamente se volverá mucho menos legible.*
