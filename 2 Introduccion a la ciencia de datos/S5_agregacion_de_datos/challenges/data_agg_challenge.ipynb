### Usando `groupby()`, encuentra cuantas peliculas tenemos registradas para cada decada

```python
peliculas.groupby((peliculas['fch_estreno'] // 10) * 10)['titulo'].count()
```

```python
peliculas_decadas = peliculas.copy()
peliculas_decadas['decadas'] = peliculas.fch_estreno.apply(lambda x: str(x)[:3]+"0's")
peliculas_decadas.groupby('decadas')['titulo'].count()
```
