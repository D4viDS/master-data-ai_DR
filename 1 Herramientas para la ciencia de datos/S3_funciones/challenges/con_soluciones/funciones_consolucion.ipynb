## 1 Escribe una función que encuentre el máximo de 3 números

```python
def max_threenumbers(a,b,c):
  if b < a > c: # b < c < a  
    print(f"El numero mayor es {a}")
  elif a < b > c: # a < c < b
    print(f"El numero mayor es {b}")
  elif a < c > b: # a < b < c
    print(f"El numero mayor es {c}")

```

```python
max_threenumbers(80,80,7)
```

## 2 Escribe una función que sume todos los números de una lista

## 3 Escribe una función que multiplique todos los números de una lista

## 4 Escribe una función que regrese la suma y la multiplicacion de todos los numeros de una lista

```python
def funcion_sumandmultiplica(*arg):
  suma = 0
  producto = 1
  for nota in arg:
        suma += nota 
  for nota in arg:
        producto *= nota 
  return suma, producto 
  
```

```python
suma, producto = funcion_sumandmultiplica(*mi_lista)
```

```python
print(f"La sumatoria de tu lista es {suma}")
print(f"La multiplicacion de tu lista es {producto}")
```

## 5 Escribe una función que revierta el orden de un string (cadena de caracteres)

```python
def inverso(palabra):
  palabra_list = list(palabra)
  palabra_revers = "".join(palabra_list[::-1])
  print(palabra_revers)
 
```

```python
inverso("Juanito")
```

## 6 Has una funcion que cheque si un numero es primo


```python
def es_primo(n):
  is_prime = True
  if n > 1:
    for i in range(2, int(n/2)):
      if (n % i) == 0:
        is_prime = False
        break
  else:
    is_prime = False
        
  return is_prime
      
    
```

```python
for i in range(100):
  if es_primo(i):
    print(i)
```

## 7 Escribe una funcion que nos devuelva el factorial de un numero

```python
def factorial(n):
  if n == 1:
    return n
  else:
    return n * factorial(n-1)
```

```python
def factorial(n):
  fact = 1
  for i in range(n):
    fact *= (i+1)
  
  return fact
```

```python
factorial(5)
```

## 8 Escribe una funcion que nos diga si una palabra es palindromo o no

```python
def palindrome(word):
  if str(word) == str(word)[::-1] :
    print("Es un palindromo")
  else:
    print("No es un palindromo")
```

```python
palindrome("madam")
```
