# List Comprehension

1.- Crear un elemento que contenga la siguiente lista numeros = [1, 2, 34, 86, 4, 5, 99, 890, 45] y solo me seleccione los numeros pares ademas de ordenarlos de menor a mayor

```python
numeros = [1, 2, 34, 86, 4, 5, 99, 890, 45]
pares = []
for num in numeros:
    if num % 2 == 0:
        pares.append(num)

pares.sort()
print(pares)
```
