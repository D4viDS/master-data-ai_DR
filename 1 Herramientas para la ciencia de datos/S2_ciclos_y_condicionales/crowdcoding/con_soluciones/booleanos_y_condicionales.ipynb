# Booleanos y condicionales

## Booleanos

Los booleanos es un tipo de datos que represantan uno de dos valores, `True` (verdadero) o `False` (falso)

```python
# De la siguiente forma los podemos asignar a una variable
x = True
y = False
print(x, y)
```

```python
# bool() convierte una variable en booleano
print(bool(1))
```

```python
# En python la mayoria de los valores son verdaderos
print(bool("abc"))
print(bool(123))
print(bool(["aguacate", "banana", "cereza"]))
```

```python
# Pero strigns vacios, el numero 0 y listas vacias son falsos
print(bool(""))
print(bool(0))
print(bool([]))
```

### Operadores de comparacion

|Operator|Name|Example|
|--- |--- |--- |
|==|igual a|x == y|
|!=|no igual a|x != y|
|>|mayor que|x > y|
|<|menor que|x < y|
|>=|igual o mayor que|x >= y|
|<=|igual o menor que|x <= y|

```python
a = 200
b = 30
c = 30

print("a == b", a == b)
print("b == c", b == c)
print("b != c", b != c)
print("a > b", a > b)
print("a < b", a < b)
print("b >= c", b >= c)
print("b <= c", b <= c)
```

### Operadores logicos

|Operator|Description|Example|
|--- |--- |--- |
|and|Verdadero si ambos valores son verdaderos|x < 5 and  x < 10| 
|or|Verdadero si uno de los valores es verdaderos|x < 5 or x < 4| 
|not|Verdadero si es falso, falso si es verdadero|not(x < 5 and x < 10)|
|&|AND|x < 5 & x < 10|
|\||OR|x < 5 \| x < 4|
|~|NOT|~(x < 5 & x < 10)|




```python
x = 7
print("x < 5 and x < 10", x < 5 and x < 10)
print("x < 5 or x < 10", x < 5 or x < 10)
print("not(x < 5)", not(x < 5))
```

### Operadores de pertenencia

|Operator|Description|Example|
|--- |--- |--- |
|in|Verdadero si una secuencia con el valor especificado esta persente en el objeto|x in y|
|not in|Verdadero si una secuencia con el valor especificado no esta persente en el objeto|x not in y|


```python
print("[0] in [0, 1, 2]" , [0] in [0, 1, 2])
print("[3] in [0, 1, 2]" , [3] in [0, 1, 2])
print("'b' in 'banana'", "b" in "banana")
print("'c' in 'banana'", "b" in "banana")
```

## Condicionales


Los booleanos y operadores booleanos son principalmente usados en condicionales `if` `elif` y `else` y en ciclos `while`

### Condicionales if

Un condicional `if` permite escribir un segmento de codigo que solo se ejecute cundo se cumpla cierta condición

```python
if True:
  print('Esto se imprime por ser verdadero')

if False:
  print('Esto no se imprime por ser falso')
```

```python
# Lo util de esto es que podemos poner expresiones booleanas
x = 5
y = 10
if x > y:
  print("x es mayor a y")

if x < y:
  print("x es menor a y")
```

Para poder usar esto, es importante aprender la sintaxis (la forma correcta de escribir) de Python

```python
# En python, se hace uso de la indentacion para indicar que cierto segmento de
#   codigo es parte de alguna condicional

if False:
  print("esto es parte del condicional if")

print("esto no lo es")
```

### Else

```python
# El operador `else` (entonces) nos permite ejecutar un pedazo de codigo si
#   y solo si el codigo dentre del operador if no se ejecuta

if False:
  print("esto no se ejecuta")
else:
  print("esto si se va a ejecutar")

if True:
  print("esto se va a ejecutar")
else:
  print("por lo que esto no se va a ejecutar")
```

### Elif

```python
# La combinacion de los condicionales else e if es elif.
# Lo que hace es que en caso de fallar una condicion if, agregar una nueva
#   condicion
if False:
  print("No se imprime")
elif True:
  print("Si se imprime")
else:
  print("Solo se imprime si el if y el elif fallan")


if False:
  print("No se imprime")
elif False:
  print("No se imprime")
else:
  print("Solo se imprime si el if y el elif fallan")
```

## Ciclos while

```python
# Los ciclos `while` nos permiten repetidamente ejecutar un segmento de codigo
#   hasta que cierta condicion sea falsa
i = 1
while i < 6:
  print(i)
  i = i +1 
```

```python
# La palabra clave `break` nos pemite salir de un ciclo
i = 1
while i < 6:
  print(i)
  if i == 3:
    print("acabamos antes de tiempo")
    break
  i = i +1 
```

```python
# La palabra clave `continue` nos permite saltarnos un bloque de codigo sin 
#   salirnos del ciclo
i = 0
while i < 6:
  i = i +1 
  if i == 3:
    print("nos saltamos un numero")
    continue
  print(i)
```

```python
# Podemos usar `else` para ejecutar un bloque de codigo una vez que la condicion
#   `while` seas falsa
while i < 6:
  print(i)
  i = i +1 
else:
  print("la condicion del while ahora es false")
```

```python
# El bloque de codigo contenido en el `else` no se ejecutara si el ciclo `while`
#   se termina con un `break`
i = 1
while i < 6:
  print(i)
  if i == 3:
    print("acabamos antes de tiempo")
    break
  i = i +1 
else:
  print("esto solo se imprime si no hay un break")
```
