```python
import random


class Pokemon:
  dano_base = 1

  def __init__(self, especie, stats, atacks, tipo, fortalezas, debilidades):
    self.especie = especie
    self.stats = stats
    self.current_stats = self.stats.copy()
    self.tipo = tipo
    self.atacks = atacks
    self.debilidades = debilidades
    self.fortalezas = fortalezas

  def centro_pokemon(self):
    self.current_stats = self.stats

  def prueba(self, rival):
    print(self.atacks[1][1])

  def pelea(self, rival):
    
    # Tu rival es furte o debil a ti?
    if self.tipo in rival.fortalezas:
      modificador_ataque = 1
      print(f"{rival.especie} es fuerte a los ataques de {self.especie} \n")
    elif self.tipo in rival.debilidades:
      modificador_ataque = 1
      print(f"{rival.especie} es debil a los ataques de {self.especie} \n")
    else:
      modificador_ataque = 1
    
    #eres fuerte o debil a tu rival?
    if rival.tipo in self.fortalezas:
      modificador_defensa = 1
      print(f"{self.especie} es fuerte a los ataques de {rival.especie} \n")
    elif rival.tipo in self.debilidades:
      modificador_defensa = 1
      print(f"{self.especie} es debil a los ataques de {rival.especie} \n")
    else:
      modificador_defensa = 1

    # quien ataca primero
    if self.current_stats["velocidad"] >= rival.current_stats["velocidad"]:
      mi_turno = True
    else:
      mi_turno = False
    
    # combate por turnos
    while (self.current_stats["hp"] > 0) & (rival.current_stats["hp"] > 0):
      if mi_turno:
        # atacando
        print("LISTA DE MOVIMIENTOS:")
        print("NUM | NOMBRE | DAÑO  | TIPO MOV")
        i = 0
        j = 1
        for i in self.atacks:
            
            print(j, self.atacks[i])
            j += 1

        seleccion = int(input("Seleccione el ataque segun el digito: "))
        if seleccion == 1:
          if self.atacks[1][2] in rival.fortalezas:
            modificador_ataque = 0.5
          elif self.atacks[1][2] in rival.debilidades:
            modificador_ataque = 1.5
          else:
            modificador_ataque = 1
          print("Usted selecciono el ataque ", self.atacks[1]) 
          dano = int(
              self.dano_base * 
              (int(self.atacks[1][1])) * 
              modificador_ataque) 
          rival.current_stats["hp"] -= dano
          print(f"{self.especie} hizo {dano} de daño a {rival.especie}")
          if self.current_stats['hp'] <= 0:
            print(f'{rival.especie} ha ganado la pelea \n')
          else:
            print(f"A {self.especie} le quedan {self.current_stats['hp']} puntos de vida")
            print(f"A {rival.especie} le quedan {rival.current_stats['hp']} puntos de vida")
        elif seleccion == 2:
          if self.atacks[2][2] in rival.fortalezas:
            modificador_ataque = 0.5
          elif self.atacks[2][2] in rival.debilidades:
            modificador_ataque = 1.5
          else:
            modificador_ataque = 1
          print("Usted selecciono el ataque ", self.atacks[2]) 
          dano = int(
              self.dano_base * 
              (int(self.atacks[2][1])) * 
              modificador_ataque) 
          rival.current_stats["hp"] -= dano
          print(f"{self.especie} hizo {dano} de daño a {rival.especie}")
          if self.current_stats['hp'] <= 0:
            print(f'{rival.especie} ha ganado la pelea \n')
          else:
            print(f"A {self.especie} le quedan {self.current_stats['hp']} puntos de vida")
            print(f"A {rival.especie} le quedan {rival.current_stats['hp']} puntos de vida")
        elif seleccion == 3:
          if self.atacks[3][2] in rival.fortalezas:
            modificador_ataque = 0.5
          elif self.atacks[3][2] in rival.debilidades:
            modificador_ataque = 1.5
          else:
            modificador_ataque = 1
          print("Usted selecciono el ataque ", self.atacks[3]) 
          dano = int(
              self.dano_base * 
              (int(self.atacks[3][1])) * 
              modificador_ataque) 
          rival.current_stats["hp"] -= dano
          print(f"{self.especie} hizo {dano} de daño a {rival.especie}")
          if self.current_stats['hp'] <= 0:
            print(f'{rival.especie} ha ganado la pelea \n')
          else:
            print(f"A {self.especie} le quedan {self.current_stats['hp']} puntos de vida")
            print(f"A {rival.especie} le quedan {rival.current_stats['hp']} puntos de vida")
        else:
          print("Seleccione un ataque valido")
        mi_turno = not mi_turno  
      else:

        atacante = random.randint(1,3)
        # defendiendo
        if atacante == 1:
          if rival.atacks[1][2] in self.fortalezas:
            modificador_defensa = 0.5
          elif rival.atacks[1][2] in self.debilidades:
            modificador_defensa = 1.5
          else:
            modificador_defensa = 1
          dano = int(
              rival.dano_base *
              (int(rival.atacks[1][1])) * 
              modificador_defensa)
          self.current_stats["hp"] -= dano
          print(f"{rival.especie} selecciono el ataque , {rival.atacks[1]}")
          print(f"{rival.especie} hizo {dano} de daño a {self.especie}")
          if self.current_stats['hp'] <= 0:
            print(f'{rival.especie} ha ganado la pelea \n')
          else:
            print(f"A {self.especie} le quedan {self.current_stats['hp']} puntos de vida")
            print(f"A {rival.especie} le quedan {rival.current_stats['hp']} puntos de vida")
        elif atacante == 2:
          if rival.atacks[2][2] in self.fortalezas:
            modificador_ataque = 0.5
          elif self.atacks[2][2] in self.debilidades:
            modificador_ataque = 1.5
          else:
            modificador_ataque = 1
          dano = int(
              rival.dano_base *
              (int(rival.atacks[2][1])) * 
              modificador_defensa)
          self.current_stats["hp"] -= dano
          print(f"{rival.especie} selecciono el ataque , {rival.atacks[2]}")
          print(f"{rival.especie} hizo {dano} de daño a {self.especie}")
          if self.current_stats['hp'] <= 0:
            print(f'{rival.especie} ha ganado la pelea \n')
          else:
            print(f"A {self.especie} le quedan {self.current_stats['hp']} puntos de vida") 
            print(f"A {rival.especie} le quedan {rival.current_stats['hp']} puntos de vida")
        elif atacante == 3:
          if rival.atacks[3][2] in self.fortalezas:
            modificador_ataque = 0.5
          elif self.atacks[3][2] in self.debilidades:
            modificador_ataque = 1.5
          else:
            modificador_ataque = 1
          dano = int(
              rival.dano_base *
              (int(rival.atacks[3][1])) * 
              modificador_defensa)
          self.current_stats["hp"] -= dano
          print(f"{rival.especie} selecciono el ataque , {rival.atacks[3]}")
          print(f"{rival.especie} hizo {dano} de daño a {self.especie}")
          if self.current_stats['hp'] <= 0:
            print(f'{rival.especie} ha ganado la pelea \n')
          else:
            print(f"A {self.especie} le quedan {self.current_stats['hp']} puntos de vida") 
            print(f"A {rival.especie} le quedan {rival.current_stats['hp']} puntos de vida")
        mi_turno = not mi_turno
    else:
      if self.current_stats["hp"] <= 0:
        print(f'{rival.especie} ha ganado la pelea \n')
      else:
        print(f'{self.especie} ha ganado la pelea \n')
 
```
